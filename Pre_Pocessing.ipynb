# Imputaciones

```python
### Valores llenados con cero
df['Tipo de vía'].fillna(0, inplace=True)
df['Número de estacionamiento'].fillna(0, inplace=True)
df['Depósitos'].fillna(0, inplace=True)
df['Método Representado'].fillna(0, inplace=True)
df['Área Terreno'].fillna(0, inplace=True)
df['Área Construcción'].fillna(0, inplace=True)
df['Estado de conservación'].fillna(0, inplace=True)
```

```python
###Valores de la media
df['Edad'].fillna(df['Edad'].mean(), inplace=True)
```

```python
Map = {}
```

```python
from numpy import nan

### Darle orden a Estado de conservación
dct = {
    0 : 0,
    'Malo': 1,
    'Regular - Malo' : 2,
    'En proyecto' : 3,
    'En construcción' : 4,
    'Regular' : 5,
    'Bueno - Regular' : 6,
    'Bueno' : 7,
    'Muy bueno' : 8
}

Map['Estado de conservación'] = dct

df['Estado de conservación'] = df['Estado de conservación'].map(dct)
```

```python
df
```

```python
### Calcular representacion de lat y lng por distrito

df['Latitud (Decimal)'].fillna(0, inplace=True)
df['Longitud (Decimal)'].fillna(0, inplace=True)

Distritos = df['Distrito'].unique()

dstr_lat = {}
dstr_lng = {}
amm_lat = {}
amm_lng = {}

for d in Distritos:
    dstr_lat[d] = 0.0
    dstr_lng[d] = 0.0
    amm_lat[d] = 0
    amm_lng[d] = 0

for index, row in df.iterrows():
    d = row['Distrito']
    if row['Latitud (Decimal)'] != 0:
        amm_lat[d] += 1
        dstr_lat[d] += row['Latitud (Decimal)']
    
    if row['Longitud (Decimal)'] != 0:
        amm_lng[d] += 1
        dstr_lng[d] += row['Longitud (Decimal)']
```

```python
for d in Distritos:
    dstr_lat[d] /= amm_lat[d]
    dstr_lng[d] /= amm_lng[d]
```

```python
for index, row in df.iterrows():
    d = row['Distrito']
    if row['Latitud (Decimal)'] == 0:
        df.at[index,'Latitud (Decimal)'] = dstr_lat[d]
    
    if row['Longitud (Decimal)'] == 0:
        df.at[index,'Longitud (Decimal)'] = dstr_lng[d]
```

```python
Map['Latitud'] = dstr_lat
Map['Longitud'] = dstr_lng
```

```python
### Mapear categoria del bien

# 0 nan
# 1 Vivienda unifamiliar = Departamento + Vivienda Unifamiliar
# 2 Terrenos = Terreno Urbano + Terreno Rústico (eriazo)
# 3 Industria = Industria + Almacén /Taller + Fundo Agrícola + Muebles y enseres + Maquinaria y/o Equipo
# 4 Negocios= Local Comercial + Oficina + Hotel
# 5 Servicios = Centro de Salud + Intitución Educativa
# 6 AVALUOS_TIPOS_INMUEBLE_VEHICULO
# 7 Estacionamiento/depósito (U.I.) 

df['Categoría del bien'].fillna(0, inplace=True)

cats = {
    0: 0,
    'Departamento' : 1,
    'Vivienda Unifamiliar' : 1,
    'Terreno Urbano' : 2,
    'Terreno Rústico (eriazo)' : 2,
    'Industria' : 3,
    'Almacén /Taller' : 3,
    'Fundo Agrícola' : 3,
    'Muebles y enseres' : 3,
    'Maquinaria y/o Equipo' : 3,
    'Local Comercial' : 4,
    'Oficina' : 4,
    'Hotel' : 4,
    'Centro de Salud' : 5,
    'Intitución Educativa' : 5,
    'AVALUOS_TIPOS_INMUEBLE_VEHICULO' : 6,
    'Estacionamiento/depósito (U.I.)' : 7
}

Map['Categoría del bien'] = cats
```

```python
df['Categoría del bien'] = df['Categoría del bien'].map(cats)
```

# Mapear categorias

```python
### Mapear Distritos

distr = df['Distrito'].unique()

dst = {}

i = 1

for d in distr:
    dst[d] = i
    i += 1

Map['Distrito'] = dst

df['Distrito'] = df['Distrito'].map(dst)
```

```python
### Mapear Departamentos

deptmnt = df['Departamento'].unique()

dpto = {}

i = 1

for d in deptmnt:
    dpto[d] = i
    i += 1

Map['Departamento'] = dpto

df['Departamento'] = df['Departamento'].map(dpto)
```

```python
### Mapear Método Representados

metodos = df['Método Representado'].unique()

mtd = {}

i = 1

for d in metodos:
    mtd[d] = i
    i += 1

Map['Método Representado'] = mtd

df['Método Representado'] = df['Método Representado'].map(mtd)
```

```python
# Cambiar fecha por año

for index, row in df.iterrows():
    fecha = row['Fecha entrega del Informe']
    df.at[index,'Fecha entrega del Informe'] = int(fecha[:4])
```

```python
for index, row in df.iterrows():
    if row['Área Terreno'] == 0:
        df.at[index,'Área Terreno'] = row['Área Construcción']

    if row['Área Construcción'] == 0:
        df.at[index,'Área Construcción'] = row['Área Terreno']
```
